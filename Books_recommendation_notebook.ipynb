```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
ps.stem("lovingly")
```

```python
# Creating a function stem to convert all the words into root words

def stem(text):
    y = []
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)    
```

```python
new_books['tag'][500]
```

```python
# Checking the stemming

stem("laura ingalls wilder juvenile fiction join the ingalls family as they celebrate little laura’s fifth birthday in the big woods of wisconsin. there are special presents from everyone, and that night laura falls asleep to the merry music of pa’s fiddle. renée graef’s enchanting full-color illustrations, inspired by garth williams’s classic artwork, bring laura and her family lovingly to life in our tenth my first little house book, adapted from laura ingalls wilder’s beloved storybooks. it’s a little house birthday to remember!")
```

```python
new_books['tag'] = new_books['tag'].apply(stem)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer(max_features =5000 , stop_words = 'english',ngram_range=(1, 2) )
```

We will now calculate the cosine distance(angular distance) not the euclidean distance(which is tip to tip).In higher dimension, euclidean distance is not a good measure.Distance is always inversally proportional to similarity.Distance less means similarity more and viceversa.

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
# Getting the cosine similarity of vectors to see how close they are

cosine_similarity(vectors)
```

```python
similarity = cosine_similarity(vectors)
```

```python
similarity[0]
```

```python
# enumerated(to keep the mapping - 1st book similarity with 2, 1st book similarity with 3 and so on in place) 
# Sorting the list of books to get the highrest similarity in descending order 
# x[1] : signifies sorting on the similarity score not the index
# getting the top 5

sorted(list(enumerate(similarity[0])),reverse = True,key = lambda x: x[1])[1:6]
```

```python
# Creating the recommend function which takes in title of the book and returns the similar books and similarity scores

def recommend_old(book):
    matches = new_books[new_books['title'] == book]
    if matches.empty:
        print(f"Book title '{book}' not found in the dataset.")
        return

    book_index = new_books[new_books['title'] == book].index[0]
    distances = similarity[book_index]
    books_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x: x[1])[1:6]
    
    for i  in books_list:
        
        #print(new_books.iloc[i[0]].title)
        score = i[1]
        title = new_books.iloc[i[0]].title 
        print(f"{title} - similarity score: {score:.3f}")
    
```

```python
# Creating the recommend function which takes in title of the book and returns the similar books and similarity scores

def recommend(book):
    matches = new_books[new_books['title'] == book]
    if matches.empty:
        print(f"Book title '{book}' not found in the dataset.")
        return

    book_index = matches.index[0]
    distances = similarity[book_index]
    books_list = sorted(list(enumerate(distances)),reverse = True,key = lambda x: x[1])[1:6]
    
    for i  in books_list:
        # Get index of recommended book
        rec_index=i[0]
        # Get title and isbn from the dataframe
        title = new_books.iloc[rec_index].title
        isbn = new_books.iloc[rec_index].isbn13
        #print(new_books.iloc[i[0]].title)
        score = i[1]
        #title = new_books.iloc[i[0]].title 
        print(f"{title},{isbn} - similarity score: {score:.3f}")
    
```
